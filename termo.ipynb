*1.102 Escreva um programa de computador que faça uma tabela de correspondência entre °C, °F, K e R, na faixa de −50 °C a 100 °C, utilizando um intervalo de 10 °C.*

```python

def fahrenheit(T_C):
    return  (T_C*5)/9 + 32 
def kelvin(T_C):
    return T_C + 273.15
def rankine (T_C):
    return (T_C*9)/5 + 491.67

for i in range(-50 , 110, 10):
    T_F = fahrenheit(i)
    T_K = kelvin(i)
    T_R = rankine(i)
    print("\n")
    print(f"A temperatura em Celsius é {i}°C")
    print(f"A temperatura em Fahrenheit é {fahrenheit(i)}°F")
    print(f"A temperatura em kelvin é {kelvin(i)}°K")
    print(f"A temperatura em Celsius é {rankine(i)}°R")

```

 *Represente graficamente a pressão atmos- férica em função da altitude (0-20 000 m) em um local em que a pressão no solo a 500 m de altitude é de 100 kPa. Use a varia- ção apresentada no Problema 1.64.*

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo os dados do problema
altitude = np.linspace(0, 20000, 100)  # Altitude de 0 a 20000 metros
pressao_solo = 100  # Pressão atmosférica no solo a 500m de altitude (kPa)
gradiente_pressao = -0.12  # Gradiente médio de pressão (kPa/m)

# Calculando a pressão atmosférica em função da altitude usando o gradiente de pressão
pressao_atmosferica = pressao_solo + gradiente_pressao * (altitude - 500)

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(altitude, pressao_atmosferica, color='blue', linewidth=2, label='Pressão Atmosférica')
plt.xlabel('Altitude (m)')
plt.ylabel('Pressão Atmosférica (kPa)')
plt.title('Pressão Atmosférica em Função da Altitude')
plt.grid(True)
plt.legend()
plt.show()
```
